```python
import numpy as np
from sklearn.datasets import *
from sklearn.svm import *
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
X, y = make_moons(n_samples = 1000, noise = 0.16)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y, cmap = 'winter' )
```

```python
line1.fit(X, y)
```

```python
clf = SVC()
```

```python
clf.fit(X, y)
```

```python
xmin, xmax = X[:, 0].min() - 1, X[:, 1].max() + 1
ymin, ymax = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(xmin, xmax, 0.01),
                    np.arange(ymin, ymax, 0.01))

grid_points = np.c_[xx.ravel(), yy.ravel()]

Z = clf.predict(grid_points)

Z = Z.reshape(xx.shape)

plt.scatter(X[:, 0], X[:, 1], c = y, cmap = 'summer', edgecolors = 'k', s = 20)
plt.contourf(xx, yy, Z, alpha = 0.3, cmap = 'winter')
plt.show()
```
